# Finding Correlation of all Features with Price of House

```python
df.corr()['price'].sort_values()
```

```python
dataset_features = df.drop(['price', 'id', 'date', 'long', 'condition', 'yr_built'], axis=1)
dataset_labels = df['price']
```

```python
dataset_features.head()
```

# Heatmap for Correaltion

```python
corr_mat = df.corr().round(2)
sns.heatmap(data=corr_mat, annot=True)
```

# Dividing dataset into Test(35%) and Training(65%) data

```python
from sklearn.model_selection import train_test_split

train_features,test_features,train_labels,test_labels = train_test_split(dataset_features,dataset_labels,test_size=0.35, random_state=2)
```

# Implementing Linear Regression

```python
from sklearn.linear_model import LinearRegression
```

```python
rg = LinearRegression()
```

```python
rg.fit(train_features, train_labels)
```

# Accuracy of model after Linear Regression

```python
rg.score(test_features,test_labels)
```

```python
predicted_price = rg.predict(test_features)
```

```python
plt.scatter(test_labels , predicted_price)
```

```python
df1 = pd.DataFrame({'Actual': test_labels, 'Predicted':predicted_price})
df2 = df1.head(10)
df2
```

# Bar graph plot b/w Actual and Predicted House Prices

# Implementing Gradient Boosting Regression

```python
from sklearn import ensemble
clf = ensemble.GradientBoostingRegressor(n_estimators = 400, max_depth = 5, min_samples_split = 2,
          learning_rate = 0.1, loss = 'ls')
```

```python
clf.fit(train_features, train_labels)
```

# Accuracy of model after Gradient Boosting Regression

```python
clf.score(test_features,test_labels)
```

```python
predicted_price1 = clf.predict(test_features)
```

```python
plt.scatter(test_labels , predicted_price1)
```
